```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cross_validation,preprocessing
from sklearn.linear_model import LinearRegression
```

```python
train = pd.read_csv("/home/shivam/Downloads/Data Science workshop/random-linear-regression/train.csv",sep=',')
train
```

```python
train['x'].head()
```

```python
# Here linear reg. is done by splitting a single train dataset into traing and testing sets.

X = np.array(train['x'])
y = np.array(train['y'])
#X = preprocessing.scale(X) # It is required when our dataset has multiple columns. At that time we need to scale it with the help of this function.
#y = np.array(train['y'])
X_train,X_test,y_train,y_test = cross_validation.train_test_split(X,y,test_size=0.2)
clf=LinearRegression()    #clf is the regression classifier
clf.fit(X_train.reshape(-1, 1),y_train)   #reshape to convert it from 1 dimentional array to 2D array.
accuracy = clf.score(X_test.reshape(-1, 1),y_test)
print accuracy
```
